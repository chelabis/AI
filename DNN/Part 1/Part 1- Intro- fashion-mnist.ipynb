```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
import keras
```

## Basic one layer/one neron network example

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])]) # Dense is layer with connected neurons (units=neurons)
model.compile(optimizer='sgd', loss='mean_squared_error')     # Define optimizer and loss function

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)   # Sample input
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)  # Sample output

model.fit(xs, ys, epochs=500)   # Fit the model with 500 iterate(epochs)
```

```python
print(model.predict(np.array([10.0, 0.0, 20.0], dtype=float)))   # Make predictions
```

## Cloth recognition 
### Using fashion mnist dataset wich includes 70000 images of 10 class clothes

```python
fashion_mnist = tf.keras.datasets.fashion_mnist            # load the data
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
print(train_labels[42])                      # print index=42 train label
plt.imshow(train_images[42], cmap='Greys')   # print index=42 train image
```

### Normalize data

```python
train_images = train_images / 255.0   #normalize pixel values of train images
test_images = test_images / 255.0     #normalize pixel values of test images
```

### Build the classification model

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),           # input layer that flatten 28 x 28 inputs to single array
    keras.layers.Dense(64, activation=tf.nn.relu),      # hidden layer with 128 neurons and relu activation function
    keras.layers.Dense(10, activation=tf.nn.softmax)     # output layer with 10 neurons for 10 classes using softmax
])
```

```python
model.compile(optimizer=tf.optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=5)
```

```python
# Evaluate the model on unseen data
model.evaluate(test_images, test_labels)
```

### Create callback

```python
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('loss')<0.5):
            print('\nLoss is low so cancelling training!')
            self.model.stop_training = True
```

```python
callbacks = myCallback()                   # add created callback class instance

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(64, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=5, callbacks=[callbacks])   # add callbacks to the model
```

## Convolutional NN

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(28,28,1)),     # first filter (1 is depth of color)
    keras.layers.MaxPooling2D(2,2),         # keep max of each 4 x 4 pixels
    keras.layers.Conv2D(64, (3,3), activation='relu'),                            # second filter
    keras.layers.MaxPooling2D(2,2),         # resize again
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
# Print the model summary
model.summary()
```

#### Because of usisng 3 x 3 conv the output is reduced by 2 pixel in x and y. (28x28) ==> (26x26)
#### Flatten layer is feed by 64 new images of 5x5 pixels which create 1600 elements which was 784 in the previous model.

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
print('Training the model:\n')
model.fit(train_images, train_labels, epochs=5)

# Evaluate on the test set
print('Testing the model:\n')
model.evaluate(test_images, test_labels)
```
